# Import Data (from CODIS)

```python
import pandas as pd 
df = pd.read_csv(r'Taiwan_Rain.csv')

df = df.drop(columns=['Unnamed: 0'])
#df_precp = df.Precp
#df = df.drop(columns=['Precp'])
df = df.dropna(how='all',axis=0)
df = df.drop(columns=['Precp','ObsTime'])
df
```

# Data Claeaning

```python
total = df.isnull().count()
null_values = df.isnull().sum()
null_percent = (df.isnull().sum() / total)

missing_data =pd.concat([total,null_values,null_percent],axis=1,keys=['total','null_values','null_percent']) 
missing_data.sort_values(by=['null_percent'],ascending=False)
```

```python
import missingno as msno
msno.bar(df)
```

```python
drop_list = list(missing_data[missing_data['null_percent']>0.7].index)
df = df.drop(drop_list,axis=1)
df.columns
```

```python
import missingno as msno
msno.bar(df)
```

```python
df_cat=[]
df_num=[]

for i in df.columns:
    if df[i].dtype == 'object':
        df_cat.append(i)
    else:
        df_num.append(i)
print(df_cat)
print(df_num)
```

```python
for i in df_cat:
    df[i] = df[i].fillna(df[i].mode()[0])
    
for i in df_num:
    df[i] = df[i].fillna(df[i].median())    
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize=(20,14))
sns.heatmap(df.corr(),annot=True,cmap='RdBu')
plt.savefig('corr.png')
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()


for i in df_cat:
    df[i] = le.fit_transform(df[i])
    
```

```python
from sklearn.model_selection import train_test_split

x = df.iloc[:,:-1]
y = df.iloc[:,-1]

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2 ,random_state = 42)
```

# RF

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV



#p={
 #   'max_depth':[10,20,50,100],
  #  'n_estimators':[100,300,500,1000] }
p={
   'max_depth':[50],
    'n_estimators':[1000] }


clf=RandomForestClassifier()
clf=GridSearchCV(clf,p,cv=10)
clf.fit(x_train,y_train)

print(clf.best_params_)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix,classification_report,roc_curve, roc_auc_score


y_pred = clf.predict(x_test)
report = classification_report(y_test,y_pred)
print(report)
print("Accuracy is :",accuracy_score(y_test,y_pred)*100,"%")
cm = confusion_matrix(y_test,y_pred)
sns.heatmap(cm,annot=True)
plt.title('Random Forest')
plt.savefig("RF_cm")
```

```python
y_pred_proba = clf.predict_proba(x_test)
pos_proba = y_pred_proba[:,1]
```

```python
fpr, tpr, thresholds = roc_curve(y_test, pos_proba)
plt.plot(fpr, tpr, '*-')
plt.plot([0, 1], [0, 1], 'r--')
#plt.legend(['Logistic regression', 'Random chance'])
plt.xlabel('FPR')
plt.ylabel('TPR')
plt.savefig('ROC')
#plt.title('ROC curve for Random Forest Classifier')
```

```python
from sklearn.metrics import roc_auc_score
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RepeatedStratifiedKFold
import numpy as np
cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)
score_val = cross_val_score(clf, x_test, y_test, scoring='roc_auc', cv=cv, n_jobs=-1)

print('val AUC: %.3f' % np.mean(score_val))
```

# SVM

```python
from sklearn.svm import SVC


clf_svm= SVC( kernel='rbf',
         class_weight='balanced',
            probability=True)

clf_svm.fit(x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix,classification_report,roc_curve, roc_auc_score


y_pred = clf_svm.predict(x_test)
report = classification_report(y_test,y_pred)
print(report)
print("Accuracy is :",accuracy_score(y_test,y_pred)*100,"%")
cm = confusion_matrix(y_test,y_pred)
sns.heatmap(cm,annot=True)
plt.title('SVM')
```

# KNN

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
clf_KNN = KNeighborsClassifier()
N={ 'n_neighbors':[i for i in range(5, 10)]}
clf_KNN = GridSearchCV(clf_KNN, N, n_jobs=-1)
clf_KNN.fit(x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix,classification_report,roc_curve, roc_auc_score


y_pred = clf_KNN.predict(x_test)
report = classification_report(y_test,y_pred)
print(report)
print("Accuracy is :",accuracy_score(y_test,y_pred)*100,"%")
cm = confusion_matrix(y_test,y_pred)
sns.heatmap(cm,annot=True)
plt.title('KNN')
```
